### 2. More Querying Data - COUNT, DISTINCT, GROUPBY, MAX, MIN SUM, AVG
This workbook expands on workbook 1, giving you more tools and keywords to query your data.

This includes use agregation functions to summarise data and get statistics about it, such as sum, average, maximum and minimum.

###### Keywords Covered - COUNT, DISTINCT, GROUPBY, MAX, MIN, SUM, AVG

```python
import sqlite3
import pandas as pd

# Connect to database
data_conn = sqlite3.connect("pokemon.db")
```

#### COUNT

Use: Returning the number of rows that match a criteria.

*Tip: Use in conjunction with other keywords such as DISTINCT, WHERE and GROUPBY for deeper insights.*

In the last workbook we used the len() python function to count the rows of data. We can get the same result using the COUNT SQL keyword.

```python
# Using len() function in python
data = data_conn.execute("SELECT * FROM pokemon").fetchall()
print(f"Rows in data: {len(data)}")

poke_count = data_conn.execute("SELECT COUNT (*) FROM pokemon").fetchall()
print("Rows in data using COUNT function:", poke_count[0][0])
```

#### DISTINCT

Use: Removing duplicates and showing only unique values.

Below we will use the DISTINCT keyword to list the types of Pokemon, and then the COUNT keyword to count them.

```python
d_types = data_conn.execute("SELECT DISTINCT type1 FROM pokemon").fetchall()
print(pd.DataFrame(d_types))

count_d_types = data_conn.execute("SELECT COUNT (DISTINCT  type1) FROM pokemon").fetchall()
print("Pokemon Types:",count_d_types[0][0])
```

#### GROUPBY 

Use: Organises data into groups based on one or more column chosen to group by.

*Tip: Use with aggregate functions like MIN, MAX, COUNT, SUM or AVG.*

Below we will use GROUP BY with the COUNT function to count how many of each Pokemon type there are.

```python
d_types = data_conn.execute("SELECT type1, COUNT(name) FROM pokemon GROUP BY type1").fetchall()
print(pd.DataFrame(d_types))
```

#### MAX

Use: Find the maximum value of a column.

Below we will find the tallest Pokemon. Note how three columns are being SELECTed.

```python
tall_pkmn = data_conn.execute("SELECT name, type1, MAX(height) FROM pokemon").fetchall()
print(pd.DataFrame(tall_pkmn))
```

#### Min

Use: Find the minimum value of a column.

Below we will find the shortest Pokemon of each type by use MIN in conjunction with the GROUP BY.

```python
short_kings = data_conn.execute("SELECT name, type1, MIN(height) FROM pokemon GROUP BY type1").fetchall()
print(pd.DataFrame(short_kings))
```

#### SUM

Use: Find the total value of a column.

Below we will sum the total of all the Pokemon IDs from 1 to 1025.

We can check this as the total of any number from 1 to n is n(n+1)/2.

So the answer shoudl be 1025(1025 + 1) / 2 = 525,825

```python
id_total = data_conn.execute("SELECT SUM(id) FROM pokemon").fetchall()
print("Total:", id_total[0][0])
```

#### AVG

Use: Calculate the mean average of a column.

*Tip: use with GROUP BY to create useful statistics.*

Below we will calculate the average wight of each Pokemon type group.

```python
avg_weight = data_conn.execute("SELECT type1, AVG(height) FROM pokemon GROUP BY type1").fetchall()
print(pd.DataFrame(avg_weight))
```
