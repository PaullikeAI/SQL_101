In this notebook we will cover the different data types used in SQL databases.

###### Keywords Covered - STRICT, INTEGER, INT, TINYINT, SMALLINT, BIGINT, REAL,FLOAT, DOUBLE, TEXT, CHAR, VARCHAR, BOOLEAN, BIT, DATE, TIME, TIMESTAMP, DATETIME, BINARY, VARBINARY, BLOB, JSON, ARRAY, ENUM, SERIAL

The database created and used in these notebooks is an SQlite database. This uses a light version of the SQL language. It uses dynamic typing. This means the types in the tables are not enforced. You can insert any type into a column unless you use the STRICT keyword as whown below.

### STRICT

Use: Enforcing types in SQlite tables.

*Tip an error will come up if you try to INSERT an incorrect data type.*

```python
import sqlite3
import pandas as pd

# Connect to database.
data_conn = sqlite3.connect("pokemon.db")
```

```python
data_conn.execute("DROP TABLE pokemon_item_price") # Delete to refresh any prior changes.

# Create the tables for the Pokémon items and their prices using the STRICT keyword.
data_conn.execute("CREATE TABLE pokemon_item_price \
                        (item_id INTEGER PRIMARY KEY AUTOINCREMENT, \
                         name TEXT NOT NULL, \
                         price INTEGER NOT NULL) STRICT")
```

<div align="center"><b>SQLite Data Types</b></div>

SQLite stores data as five storage classes, but we will explore the others in SQL for a wider picture of SQL databases. The five SQLite data types are shown below.

| Category | Keyword | Use Case |
|:---------------|:----------------------------------|:---------------------------|
| Whole numbers | INTEGER |Integer values. |
| Decimal numbers |REAL |Floating point numbers. |
| Text | TEXT |Any text or strings. |
| N/A |NULL |N/A values. |
| Binary data | BLOB | Images and files. |
