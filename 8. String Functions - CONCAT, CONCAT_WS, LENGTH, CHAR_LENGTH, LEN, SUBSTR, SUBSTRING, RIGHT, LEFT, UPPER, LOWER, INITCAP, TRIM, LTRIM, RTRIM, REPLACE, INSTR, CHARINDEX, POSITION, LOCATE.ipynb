## 8. String Functions: CONCAT, CONCAT_WS, LENGTH, CHAR_LENGTH, LEN, SUBSTR, SUBSTRING, RIGHT, LEFT, UPPER, LOWER, INITCAP, TRIM, LTRIM, RTRIM, REPLACE, INSTR, CHARINDEX, POSITION, LOCATE

In this notebook we will cover functions to modify, search and measure strings in various ways..

Also the use of triple quotations *""" """* in writing SQL and the Pandas *read_sql* function.

###### Keywords Covered - CONCAT, CONCAT_WS, LENGTH, CHAR_LENGTH, LEN, SUBSTR, SUBSTRING, RIGHT, LEFT, UPPER, LOWER, INITCAP, TRIM, LTRIM, RTRIM, REPLACE, INSTR, CHARINDEX, POSITION, LOCATE

```python
import sqlite3
import pandas as pd

# Connect to database
data_conn = sqlite3.connect("pokemon.db")

data_conn.execute("DROP TABLE pokemon_items") # Delete to refresh any prior changes.

data_conn.execute("CREATE TABLE pokemon_items \
                        (name TEXT NOT NULL, \
                         effect TEXT)")

data_conn.execute("INSERT INTO pokemon_items (name, effect) VALUES \
                ('Poké Ball', 'Catches Pokémon.'), ('great Ball', 'Catches Pokémon better.'), ('Ultra Ball', 'Catches Pokémon great!'), \
                ('potion', NULL), ('Super Potion', 'Heals 50 HP.'), ('Full Restore', NULL), \
                ('revive', 'Revives your Pokémon.'), ('Escape Rope', 'Takes you to the entrance of a cave.')")
```

#### CONCAT

Use: Concatenation of two or more strings.

*Tip: Can also use double pipe operator || or + symbol. Double pipe operator returns NULL if you attempt to concatonate any NULL values.*

#### """ """


Use: You don't have to use \ at the end of each line.

```python
df = pd.DataFrame(data_conn.execute("""SELECT CONCAT(name, ": ", effect)
                                       FROM pokemon_items""").fetchall())
df.columns = ["CONCAT(name, ': ', effect)"]
df
```

We can do the same thing by using the ***read_sql(sql=[SQL], con=[Connection])*** Pandas function. This function passes the column names straight into the dataframe, skipping the process of adding the column names as we were doing before.

```python
pd.read_sql(sql="""SELECT name || ': ' || effect 
                   FROM pokemon_items""", con=data_conn)
```

#### CONCAT_WS

Use: Concatonation with a seperator.

*Tip: Skips NULL values automatically like CONCAT.*

Below we will use ": " as the seperator, instead of it being an additional concatonation term.
This is more useful for longer lists.

```python
pd.read_sql(sql="""SELECT CONCAT_WS(": ", name,  effect) 
                   FROM pokemon_items""", con=data_conn)
```

#### LENGTH (CHAR_LENGTH or LEN in other SQL varieties)

Use: Tell a user the length of a string.

```python
pd.read_sql(sql="""SELECT name, LENGTH(name)
                   FROM pokemon_items""", con=data_conn)
```

#### SUBSTR SUBSTRING

Use: Extracting a portion of a string.

*Syntax: SUBSTR(string, start, length) (same for SUBSTRING).*

*Tip: Start is 1, not 0 for the first character in a string.*

RIGHT and left are in not in SQLite, but in other variants. You can make an equivelant using SUBSTR and LENGTH (to see how long the string is) in SQLite.

#### RIGHT

Use: Selecting some of the characters form the Right of a a string.

*Syntax: RIGHT(string, length).*

#### LEFT

Use: Selecting some of the characters form the Left of a a string.

*Syntax: LEFT(string, length).*

```python
pd.read_sql(sql="""SELECT name, SUBSTR(name, 1, 4), SUBSTRING(name, 5, 10) 
                   FROM pokemon_items""", con=data_conn)
```

#### UPPER

Use: Makes the whole string uppercase.

#### LOWER

Use: Makes whole string lowerrcase.

#### INITCAP (not in SQLite)

Use: Capitalises the first letter of every word in a string and converts all other letters to lowercase

See below where we make the first letter of each string uppercase. Useful for name or company fields.

```python
pd.read_sql(sql="""SELECT name, UPPER(name),
                                LOWER(name),
                                CONCAT(UPPER(SUBSTR(name, 1, 1)), LOWER(SUBSTR(name, 2, 151)))
                                FROM pokemon_items""", con=data_conn)
```

#### TRIM

Use: Trimming white space or characters from both sides of a string. 

*Syntax: Use TRIM(string, char) to delete a specific character (or characters) - Same with LTRIM and RTRIM.*

#### LTRIM

Use: Trimming white space or characters from the left side of a string. 

#### RTRIM

Use: Trimming white space or characters from the right side of a string. 

Below we will demonstrate all three, trimming firstly just whitespace, then specific characters.

```python
pd.read_sql(sql="""SELECT TRIM('   Ash     '),
                   LTRIM('.!.Misty', '.!'),
                   RTRIM('Brock!!!', '!')""", con=data_conn)
```

#### REPLACE

Use: Replaces all occurences of a substring with another string.

*Tip: Learn it well as it is very useful for data cleaning.*

```python
# Replacing Poké with Digi.
pd.read_sql(sql="""SELECT name, REPLACE(name, 'Poké', 'Digi'),
                                REPLACE(effect, 'Poké', 'Digi')
                                FROM pokemon_items""", con=data_conn)
```

#### INSTR 

Use:  Finding the position of a substring (count starts at 1 for first letter).

*Syntax: INSTR(string, substring)*

*Tip: Returns NULL if string or substring are NULL*

#### CHARINDEX POSITION LOCATE 

Use: Same as INSTR but used in MYSQL and other SQL database languages.

These substring tools can be useful to to parse strings by finding delimiters.

Below we will use INSTR to find the spaces to get the first word of each name.

Note, we add a space to name [INSTR(name || ' '] to find a space at the end of the one word items, else INSTR would return 0, as they have no spaces.

```python
# Replacing Poké with Digi.
pd.read_sql(sql="""SELECT name,
                        INSTR(name, ' '),
                        SUBSTR(name, 1, INSTR(name || ' ', ' ')-1)
                        FROM pokemon_items""", con=data_conn)
```
