### 1. Querying Data - SELECT, FROM, WHERE, ORDER BY, ASC, DESC

Queries are an essential tool in SQL. This workbook aims to give you the tools you need to be able to query data and obtain the information you want.

###### Keywords Covered: SELECT, FROM, PRAGMA, WHERE, ORDER BY, ASC, DESC

```python
import sqlite3
import pandas as pd

# Connect to database
data_conn = sqlite3.connect("pokemon.db")
```

#### SELECT

Use: Specifying which columns you want to retrieve.

*Tip: Use * to select all columns. Multiple columns can be selected by seperating them with commas.*

#### FROM

Use: Specifies the table that you want to get the data from.

Below we will SELECT all columns from the pokemon table.

```python
data = data_conn.execute("SELECT * FROM pokemon").fetchall()

print(f"Rows in data: {len(data)}")
print(data[0:3]) # Print first 3 rows.
```

```python
# Display in a cleaner format by putting the data into a dataframe.
df = pd.DataFrame(data)
df.head(3)
```

This is a large amount of data, 1025 rows, to view all at once. We need to filter it to selectively to view relevant parts. To do this we first need to view the column names in the pokemon table using the PRAGMA keyword to know what we can filter by.

#### PRAGMA

Use: Query internal database information like column names.

*Note: This keyword is unique to SQLite.*

```python
# Retrieve column data.
cols = data_conn.execute("PRAGMA table_info(pokemon)").fetchall()

for col in cols:
    print(col)
```

```python
# This can also be used to list the tables in the database.
tables = data_conn.execute("PRAGMA table_list").fetchall()

for table in tables:
    print(table)
```

'sqlite_schema', sqlite_temp_schema' and 'sqlite_sequence are the default SQLite tables that contain the structure, temporary objects and autoincrement values. pokemon, votes and users are tables that store data.

### Filtering Queries

Now we know the column names for the pokemon table are: 'id', 'name', 'type1', 'type2', 'height', 'weight', 'dex_entry', 'image_url', 'elo', 'wins' and 'losses'.

Say we want to know how many type1 Water Pokemon there are. We can do this using the WHERE keyword.

#### WHERE

Use: Filtering queries by a specified constraint.

```python
water_pkmn = data_conn.execute("SELECT * FROM pokemon WHERE type1 = 'water'").fetchall()

print(f"Water Pokemon in data: {len(water_pkmn)}")
```

WHERE can be used with conjunctions like AND to filter in more than one way.

#### AND

Use: Filtering records based on more than one condition.

For instance, if we wanted know not just how many water Pokeone there were in total, but how many there are in the original 151 Pokemon.

```python
water_pkmn = data_conn.execute("SELECT * FROM pokemon WHERE type1 = 'water' and id < 152").fetchall()

print(f"Water Pokemon in first 151 Pokemon: {len(water_pkmn)}")
```

The next Keyword we could use in a query is ORDER BY.

#### ORDER BY

Use: Ordering data by a certain column in ascending or descending order

#### ASC

Use: Specifying the data to be sorted in ascending order (smallest to largest).

#### DESC

Use: Specifying the data to be sorted in descending order (largest to smallest).

In the below example, we will list the Pokemon in reverse order by listing by their id in descending order.

```python
desc_pkmn = data_conn.execute("SELECT * FROM pokemon ORDER BY id DESC").fetchall()
pd.DataFrame(desc_pkmn).head(3)
```

Lastly, some notes on the fetchall() function we have been using. As well as this there are two other variants, fetchone() that returns one row, or fetchmany(), which returns a list of tuples if you want results in batches.


|Method | What it returns | When to use |
| --- | --- | --- |
|.fetchall() | List of tuples (all rows) | When you want all results at once |
|.fetchone() |Single tuple (or None) | When you expect only 1 row |
|.fetchmany(size) | List of tuples (up to size) | When you want results in batches |

```python
# As shown, using fetchone() will only fetch the first result of a query.
one_pkmn = data_conn.execute("SELECT * FROM pokemon").fetchone()
print(one_pkmn)
```

```python
# Using fetchmany() will fetch the number of rows specified. In this case the first three electric Pokemon.
three_pkmn = data_conn.execute("SELECT * FROM pokemon WHERE type1 = 'electric'").fetchmany(3)
pd.DataFrame(three_pkmn).head(3)
```
