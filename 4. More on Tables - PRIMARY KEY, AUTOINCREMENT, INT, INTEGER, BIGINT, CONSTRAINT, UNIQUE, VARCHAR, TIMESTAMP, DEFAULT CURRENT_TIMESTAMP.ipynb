## 4. More on Tables: PRIMARY KEY, AUTOINCREMENT, INT, INTEGER, BIGINT, CONSTRAINT, UNIQUE, VARCHAR, TIMESTAMP, DEFAULT CURRENT_TIMESTAMP

In this notebook we will expand your knowledge on tables and keys. Keys are contraints that are used to enforce data integretity, uniqueness and relationships between tables. They are essential when working with and designing databases.

Also covered are few of the datatypes, how to add timestamps to your table and record when a row was added.


###### Keywords Covered - PRIMARY KEY, AUTOINCREMENT, INT, INTEGER, BIGINT, CONSTRAINT, UNIQUE, VARCHAR, TIMESTAMP, DEFAULT CURRENT_TIMESTAMP,

```python
import sqlite3
import pandas as pd

# Connect to database
data_conn = sqlite3.connect("pokemon.db")
```

#### PRIMARY KEY

Use: Creaing a unique refernce for each row in a table. It must be unique accross all rows and connot contain NULL values. 

*Tip: You can only have one PRIMARY KEY per table.*

*Tip: A PRIMARY KEY can be made from one or several columns as shown below.*

#### AUTOINCREMENT

Use: This automatically generates a PRIMARY KEY, starting at 1 and increasing by one per row.

*Tip: You can change the amount incremented by. eg.g  SET @@auto_increment_increment = 10*


### INT or INTEGER

A 32 bit integer.

Use: Keys, IDs, counting, any integer value.

### BIGINT

A 64 bit integer.

Use: Large IDs, timestamps.

*Tip: Use if you are going to add millions of rows to the table.*

```python
# Create a Primary Key using a single column and AUTOINCREMENT
data_conn.execute("CREATE TABLE employees (employeeID INTEGER PRIMARY KEY AUTOINCREMENT, \
                                           name TEXT, \
                                           phone_number INTEGER)")
```

```python
# Create a Primary KEY using multiple columns. This is a Composite Primary Key
data_conn.execute("CREATE TABLE shops (ShopID INT, \
                                    CountyID INT, \
                                    Shop_name TEXT, \
                                    PRIMARY KEY (ShopID, CountyID))")
```

### CONSTRAINT

Using CONSTRAINT means combination primary keys will be given your chosen name in the schema rather than a random one. The CONSTRAINT can also be removed at a later date.

Use: Use to make combination Primary keys in a more professional manner.

```python
# Using CONSTRAINT to make the Primary Key, along with NOT NULL to enforce the IDs used in the key.
data_conn.execute("CREATE TABLE shops_c (ShopID INT NOT NULL, \
                                         CountyID INT NOT NULL, \
                                         Shop_name TEXT, \
                                         CONSTRAINT pk_shops PRIMARY KEY (ShopID, CountyID))")
```

### UNIQUE

Use: This keyword enforces uniqueness in a column (or multiple coulmns), for a any type of value.
This can be a NULL value, unlike with a PRIMARY KEY.

*Tip: For best practise use it in conjunction with CONSTRAINT.*
*This means any errors will mention the CONSTRAINT*

### VARCHAR

Use: Use this to set a max length for a string field.

```python
# Creating a table with the name being a VARCHAR that has an enforceed UNIQUE name using CONSTRAINT.
data_conn.execute("CREATE TABLE employees_uq (employeeID INTEGER PRIMARY KEY AUTOINCREMENT, \
                                              name VARCHAR(20) NOT NULL, \
                                              CONSTRAINT uq_name UNIQUE (name))")
```

### TIMESTAMP

Use: This is a datatype that is used to store data and time values.

### DEFAULT CURRENT_TIMESTAMP

Use: Automatically inserting the current date and time when a new row is added.

```python
# Creating a table where the created_at column automatically stores the date and time.
data_conn.execute("CREATE TABLE employee_breaks(breakID INTEGER PRIMARY KEY NOT NULL, \
                                                created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP)")
```

```python
# Dropping demo tables.
data_conn.execute("DROP TABLE employees")
data_conn.execute("DROP TABLE shops")
data_conn.execute("DROP TABLE shops_c")
data_conn.execute("DROP TABLE employees_uq")
data_conn.execute("DROP TABLE employee_breaks")
```
